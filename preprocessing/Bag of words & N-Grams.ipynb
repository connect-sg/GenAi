```python
import pandas as pd
import chardet

messages = pd.read_csv("spam.csv", encoding='ISO-8859-1', names = ['label','messages', 'NA1','NA2','NA3'], skiprows =1)
```

```python
messages = messages.iloc[:,0:2]
```

```python
messages
```

```python
## Data cleaning and pre-processing

import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0,len(messages)):
    # replace any character apart from letters to a blank
    review = re.sub('[^a-zA-Z]', ' ', messages['messages'][i])
    
    # to lower cawe
    review = review.lower()
    
    # split into words
    review = review.split()
    
    # remove stopwords
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## create bag of words model

from sklearn.feature_extraction.text import CountVectorizer

# take top 2500 words with max frequency
cv = CountVectorizer(max_features = 2500)
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
X # this is not binary bag of words
  # can have values apart from 1s and 0s
```

```python
## create bag of words model

from sklearn.feature_extraction.text import CountVectorizer

# take top 2500 words with max frequency and make it binary
cv = CountVectorizer(max_features = 2500, binary = True)
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
X # now it will have only 1s and 0s
```

```python
# list of top words with indexes of columns
cv.vocabulary_
```

```python
## create bag of words model with N-gram

from sklearn.feature_extraction.text import CountVectorizer

# take top 2500 words with max frequency
cv = CountVectorizer(max_features = 2500, ngram_range = (1,1))# all feature will be 1 i.e. uni-gram
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
X
```

```python
## create bag of words model with N-gram

from sklearn.feature_extraction.text import CountVectorizer

# take top 2500 words with max frequency
cv = CountVectorizer(max_features = 2500, ngram_range = (1,2))# all feature will be 1 & 2 i.e. uni-gram and bi-gram
```

```python
X = cv.fit_transform(corpus).toarray()
```
